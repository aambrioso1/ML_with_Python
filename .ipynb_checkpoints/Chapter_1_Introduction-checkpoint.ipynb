## 1.7.2 Measuring Success:  Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
print("feature names:", iris_dataset['feature_names'])
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

## 1.7.4: Building Your First Model:  k-Nearest Neighbors

Some documentation on this classifier:

https://scikit-learn.org/stable/modules/neighbors.html#neighbors

From SciKit Learn:
The principle behind nearest neighbor methods is to find a predefined number of training samples closest in distance to the new point, and predict the label from these. The number of samples can be a user-defined constant (k-nearest neighbor learning), or vary based on the local density of points (radius-based neighbor learning). The distance can, in general, be any metric measure: standard Euclidean distance is the most common choice. Neighbors-based methods are known as non-generalizing machine learning methods, since they simply remember all of its training data (possibly transformed into a fast indexing structure such as a Ball Tree or KD Tree).

https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

https://en.wikipedia.org/wiki/K-nearest_neighbors_algorithm

The curse of dimensionality:

https://en.wikipedia.org/wiki/Curse_of_dimensionality#Machine_Learning

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

## 1.7.5: Making Predictions

```python
X_new = np.array([[6.7, 3., 5.2, 2.3]])
print(f'X_new.shape: {X_new.shape}')
prediction = knn.predict(X_new)
print(f'Prediction: {prediction}')
print(iris_dataset['target_names'][prediction])
```

## 1.7.6: Evaluating the Model

```python
# I iterate through the test and prediction data to see which predictions were correct.
y_pred = knn.predict(X_test)
for i in range(len(X_test)):
    test = y_pred[i] == y_test[i]
    print(f'X_test[{i}]: {X_test[i]}, y_test[{i}]: {y_test[i]}, test: {test}')

print(f'Test set predictions:\n{y_pred}')
print(f'Test set score: {np.mean(y_pred == y_test):.2f}')
print(f'Test set score: {knn.score(X_test, y_test):.2f}')
```

Use knn to train a predictor to identify the quadrant a point is in.   First we create some training data.

```python
import random as rd
n = 100
pairs = []
for i in range(n):
    x = rd.randint(-10,10)
    y = rd.randint(-10,10)
    if x != 0 and y != 0:
        pairs.append([x,y])

# Remove points that are on the axis.
for pair in pairs:
    x, y = pair
    if x == 0 or y == 0:
        pairs.remove(pair)

quadrant = []
for pair in pairs:
    x, y = pair
    if x > 0 and y > 0:
        quadrant.append(1)
    elif x < 0 and y > 0:
        quadrant.append(2)
    elif x < 0 and y < 0:
        quadrant.append(3)
    elif x > 0 and y < 0: 
        quadrant.append(4)
        
print(pairs, quadrant, len(pairs), len(quadrant))
```
